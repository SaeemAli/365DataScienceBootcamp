## Predict absenteeism from work
The target variable is 'Absenteeism Time in Hours'. The rest of the columns will be used to make the prediction

### Imports

```python
import pandas as pd
```

### Get data

```python
raw_csv_data = pd.read_csv('Absenteeism_data.csv')
```

```python
raw_csv_data.head()
```

```python
df = raw_csv_data.copy()
```

```python
df.info() # Check for missing values
```

### Drop ID

```python
# The ID doesn't help in our analysis so we can remove it

df = df.drop(['ID'], axis = 1) # axis = 1 tells the program we are remove the column heading and not a row of data
df.head()
```

### Reason for Absence

```python
print("Min: ", df['Reason for Absence'].min())
print("Max: ", df['Reason for Absence'].max())
```

```python
sorted(df['Reason for Absence'].unique()) # The number 20 is not present in the reasons for absence
# The values here are categorical and not numerical
```

### Dummy Variable

```python
reason_columns = pd.get_dummies(df['Reason for Absence'], dtype=int, drop_first = True) # Reason 0 is removed
```

```python
pd.set_option('display.max_columns', None)
reason_columns
```

### Group reasons together

Reasons 1-14: dieseases <br>
Reasons 15-17: pregency <br>
Reasons 18-21: poisoning or uncategorised signs <br>
Reasons 22-28: Light reasons (appointments)

```python
reason_columns.columns.values
```

```python
df = df.drop(['Reason for Absence'], axis = 1)
```

```python
reason_columns.loc[:, 1:14].max(axis = 1) # For each row, it shows whether the reason was any number between 1 and 14
```

```python
reason_type_1 = reason_columns.loc[:, 1:14].max(axis = 1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis = 1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis = 1)
reason_type_4 = reason_columns.loc[:, 22:].max(axis = 1)
```

### Concatenate reasons to DF

#### Do not run the cell below multiple times!

```python
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis = 1)
```

```python
df.head(10)
```

```python
df.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
df.columns = column_names
```

```python
df.head()
```

### Reorder Columns

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df = df[column_names_reordered]
df.head()
```

### Create Checkpoint

```python
df_reason_mod = df.copy() # Checkpoint after processing the reason column
```

## Date

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
```

```python
df_reason_mod['Date'] # Format the date into Year - Month - Day
```

```python
df_reason_mod.head()
```

#### Extract Month

```python
df_reason_mod['Date'][0].month
```

```python
list_months = []
```

```python
for i in range(700):
    list_months.append(df_reason_mod['Date'][i].month)
```

```python
list_months
```

```python
df_reason_mod['Month Value'] = list_months
```

```python
df_reason_mod.head(10)
```

### Extract Day
Days go from 0-6

```python
df_reason_mod['Date'][699].weekday()
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(date_to_weekday)
```

```python
df_reason_mod.head()
```

### Education

```python
df_reason_mod['Education'].unique()
```

```python
df_reason_mod['Education'] = df_reason_mod['Education'].map({1: 0, 2: 1, 3: 1, 4: 1})
```
